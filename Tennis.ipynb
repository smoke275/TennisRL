### 2. Examine the State and Action Spaces

Run the code cell below to print some information about the environment.

````python
# reset the environment
env_info = env.reset(train_mode=True)[brain_name]

# number of agents 
num_agents = len(env_info.agents)
print('Number of agents:', num_agents)

# size of each action
action_size = brain.vector_action_space_size
print('Size of each action:', action_size)

# examine the state space 
states = env_info.vector_observations
state_size = states.shape[1]
print('There are {} agents. Each observes a state with length: {}'.format(states.shape[0], state_size))
print('The state for the first agent looks like:', states[0])
````

### 3. Take Random Actions in the Environment

In the next code cell, you will learn how to use the Python API to control the agent and receive feedback from the environment.

for i in range(5):                                         # play game for 5 episodes

````python
import torch
from maddpg_agent import Agent
from collections import deque
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
````

````python
agent = Agent(state_size=state_size, action_size=action_size, num_agents=num_agents, random_seed=1)

def maddpg(n_episodes=6000, max_t=300, print_every=100):
    scores_deque = deque(maxlen=print_every)
    scores = []
    mean_scores = []   
    
    for i_episode in range(1, n_episodes+1):
        env_info = env.reset(train_mode=True)[brain_name]
        state = env_info.vector_observations
        agent.reset()
        score = np.zeros(num_agents)
        for t in range(max_t):
            action1 = agent.act(state[0])
            action2 = agent.act(state[1])
            action = np.concatenate((action1, action2), axis=0)
            action = np.clip(action, -1, 1)
            env_info = env.step(action)[brain_name]
            next_state = env_info.vector_observations
            reward = env_info.rewards
            done = env_info.local_done
            agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            if np.any(done):
                break
        scores_deque.append(np.max(score))
        scores.append(np.max(score))    
        mean_scores.append(np.mean(scores_deque))  # save running mean
        print('\rEpisode {}\tAverage Score: {:.4f}'.format(i_episode, np.mean(scores_deque)), end="")
        torch.save(agent.actor_local.state_dict(), 'checkpoint_actor.pth')
        torch.save(agent.critic_local.state_dict(), 'checkpoint_critic.pth')
        if i_episode % print_every == 0:
            print('\rEpisode {}\tAverage Score: {:.4f}'.format(i_episode, np.mean(scores_deque)))
        if np.mean(scores_deque) > 0.5:
            torch.save(agent.actor_local.state_dict(), 'solved_checkpoint_actor.pth')
            torch.save(agent.critic_local.state_dict(), 'solved_checkpoint_critic.pth')
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.4f}'.format(i_episode-100,
                                                                                         np.mean(scores_deque)))
            break

    return scores, mean_scores

scores, mean_scores = maddpg()
````

````python
import matplotlib.pyplot as plt
import numpy as np

def plot_training_results(scores, mean_scores, save_path=None):
    """
    Create an enhanced plot of MADDPG training results using only matplotlib.
    
    Args:
        scores: List of episode scores
        mean_scores: List of running mean scores
        save_path: Optional path to save the plot
    """
    # Create figure with larger size for better visibility
    fig, ax = plt.subplots(figsize=(12, 8))
    
    episodes = np.arange(1, len(scores) + 1)
    mean_episodes = np.arange(1, len(mean_scores) + 1)
    
    # Simple smoothing using moving average (no external dependencies)
    def simple_smooth(data, window=10):
        if len(data) < window:
            return data
        smoothed = []
        for i in range(len(data)):
            start = max(0, i - window // 2)
            end = min(len(data), i + window // 2 + 1)
            smoothed.append(np.mean(data[start:end]))
        return smoothed
    
    # Plot raw scores with transparency
    ax.plot(episodes, scores, alpha=0.4, color='lightblue', linewidth=1, 
            label='Episode Scores')
    
    # Plot smoothed scores if we have enough data
    if len(scores) > 20:
        smoothed_scores = simple_smooth(scores, window=20)
        ax.plot(episodes, smoothed_scores, color='blue', linewidth=2, 
                label='Smoothed Scores')
    
    # Plot running mean with emphasis
    ax.plot(mean_episodes, mean_scores, color='red', linewidth=3, 
            label='Running Mean (100 episodes)')
    
    # Add target line (adjust target_score for your environment)
    target_score = 0.5  # Change this to your environment's target
    ax.axhline(y=target_score, color='green', linestyle='--', linewidth=2, 
               alpha=0.8, label=f'Target Score ({target_score})')
    
    # Customize the plot
    ax.set_xlabel('Episode', fontsize=14, fontweight='bold')
    ax.set_ylabel('Score', fontsize=14, fontweight='bold')
    ax.set_title('MADDPG Training Progress', fontsize=16, fontweight='bold', pad=20)
    
    # Add grid with custom styling
    ax.grid(True, alpha=0.3, linestyle='-', linewidth=0.5, color='gray')
    
    # Customize legend
    legend = ax.legend(loc='upper left', frameon=True, fancybox=True, shadow=True)
    legend.get_frame().set_alpha(0.9)
    
    # Add statistics text box
    if len(scores) > 0:
        max_score = np.max(scores)
        min_score = np.min(scores)
        final_mean = mean_scores[-1] if len(mean_scores) > 0 else 0
        last_100_mean = np.mean(scores[-100:]) if len(scores) >= 100 else np.mean(scores)
        
        stats_text = f'Max Score: {max_score:.3f}\n'
        stats_text += f'Min Score: {min_score:.3f}\n'
        stats_text += f'Final Mean: {final_mean:.3f}\n'
        stats_text += f'Last 100 Avg: {last_100_mean:.3f}\n'
        stats_text += f'Total Episodes: {len(scores)}'
        
        # Create text box with border
        props = dict(boxstyle='round,pad=0.5', facecolor='lightgray', alpha=0.8, 
                    edgecolor='black', linewidth=1)
        ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, fontsize=10,
                verticalalignment='top', bbox=props)
    
    # Set axis limits
    ax.set_xlim(1, len(scores))
    
    # Improve tick labels
    ax.tick_params(axis='both', which='major', labelsize=11)
    
    # Add some style improvements
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(0.5)
    ax.spines['bottom'].set_linewidth(0.5)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    plt.show()

def plot_simple_comparison(scores_list, labels, save_path=None):
    """
    Compare multiple training runs using only matplotlib.
    
    Args:
        scores_list: List of score arrays from different runs
        labels: List of labels for each run
        save_path: Optional path to save the plot
    """
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Predefined colors for different runs
    colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink']
    linestyles = ['-', '--', '-.', ':', '-', '--', '-.']
    
    for i, (scores, label) in enumerate(zip(scores_list, labels)):
        episodes = np.arange(1, len(scores) + 1)
        color = colors[i % len(colors)]
        linestyle = linestyles[i % len(linestyles)]
        
        ax.plot(episodes, scores, color=color, linewidth=2, alpha=0.8, 
               linestyle=linestyle, label=label)
    
    ax.set_xlabel('Episode', fontsize=14, fontweight='bold')
    ax.set_ylabel('Score', fontsize=14, fontweight='bold')
    ax.set_title('MADDPG Training Comparison', fontsize=16, fontweight='bold', pad=20)
    
    ax.grid(True, alpha=0.3)
    ax.legend(loc='best', frameon=True, fancybox=True, shadow=True)
    
    # Style improvements
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    plt.show()

plot_training_results(scores, mean_scores)
````
